```python
import pandas as pd

data = pd.read_csv('165 data Sheet2.csv')
data = data.drop(columns  = {'mean'}) #focusing on UB and LB
data
```

We are going to use Normal Distribution to approximate the sales.

```python
import numpy as np

normal_draws = []
for _ in range(10000):
    sampled_index = np.random.choice(data.index, p=data['weight'])
    lb = data.loc[sampled_index, 'LB']
    ub = data.loc[sampled_index, 'UB']
    mean = (ub + lb) / 2
    sd = (ub - lb) / 2 / 1.28
    normal_draw = np.random.normal(loc=mean, scale=sd)
    normal_draws.append(normal_draw)

confidence_interval_normal = np.percentile(normal_draws, [10, 90])
confidence_interval_normal
```
